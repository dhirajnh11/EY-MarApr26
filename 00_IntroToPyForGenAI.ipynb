List Comprehension

```python
age = [10, 20, 30 , 40]
print('old values')
print(age)
# objective is to double them up

# traditional way
counter = 0
for a in age:
  newval = a * 2
  age[counter] = newval
  counter = counter + 1

print('new age')
print(age)
```

```python
age = [10, 20, 30 , 40]
print('old values')
print(age)
# objective is to double them up

# list comprehension way
age = [ a*2 for a in age]

print('new age')
print(age)
```

```python
# dictionary comprehension is also possible

sample_dictionary = { "kids": "policy 1", "adults": "policy 2"}

# you wanted to reverse the structure, and map policies to age group instead

sample_dictionary = { value:key for key,value in sample_dictionary.items() }

sample_dictionary
```
